```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_l, train.y, test_size=0.33)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
   models = {}

# Lasso Regression
from sklearn.linear_model import Lasso
models["LASSO"] = Lasso()

# ridge
from sklearn.linear_model import Ridge
models["ridge"] =Ridge()

# Decision Trees
from sklearn.tree import DecisionTreeRegressor
models["DCTREE"] = DecisionTreeRegressor()

# Random Forest
from sklearn.ensemble import RandomForestRegressor
models["RFR"] = RandomForestRegressor()

# GBR
from sklearn.ensemble import GradientBoostingRegressor
models["GBR"] = GradientBoostingRegressor()

# Light gbr
from lightgbm import LGBMRegressor
models["LGBR"] = LGBMRegressor()

#ADA BOOST
from sklearn.ensemble import AdaBoostRegressor
models["ADABOOST"] = AdaBoostRegressor()
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
r2_scores={}
mean_absolute_errors={}
RMSEs={}
for key in models.keys():
    # Fit the classifier
    models[key].fit(X_train, y_train)
    # Make predictions
    predictions = models[key].predict(X_test)
    # Calculate metrics
    r2_scores[key] = r2_score(predictions, y_test)
    mean_absolute_errors[key] = mean_absolute_error(predictions, y_test)
    RMSEs[key] = np.sqrt(mean_absolute_error(predictions, y_test))
```

```python
df_model = pd.DataFrame(index=models.keys(), columns=["r2_scores","mean_absolute_errors","RMSEs"])
df_model['r2_scores'] = r2_scores.values()
df_model['mean_absolute_errors'] = mean_absolute_errors.values()
df_model['RMSEs'] = RMSEs.values()

df_model
```
