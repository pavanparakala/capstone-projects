```python
#Apply label encoder.
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
cat_cols.shape,cat_cols_test.shape
```

```python
com=pd.concat([cat_cols,cat_cols_test],axis="rows")
```

```python
com.shape
```

```python
com=com.apply(label.fit_transform)
```

```python
cat_cols_clean=com.iloc[0:4209,:]
cat_cols_test_clean=com.iloc[4209:,]
```

```python
cat_cols_clean.shape,cat_cols_test_clean.shape
```

```python
train=pd.concat([cat_cols_clean,num_cols],axis="columns")
test=pd.concat([cat_cols_test_clean,num_cols_test],axis="columns")
```

```python
train.shape,test.shape
```

```python
#Perform dimensionality reduction.
```

```python
#because we have target variable y i choose LDA
```

```python
target=train.y
train_t=train.drop(["y"],axis="columns")
```

```python
target=target.astype("int")
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda=LinearDiscriminantAnalysis()
```

```python
lda.fit(train_t,target)
```

```python
lda_train=lda.transform(train_t)
```

```python
lda_train=pd.DataFrame(lda_train)
```

```python
lda_train
```

```python
lda_test=lda.transform(test)
```

```python
lda_test=pd.DataFrame(lda_test)
```

```python
lda_test
```

```python
#Predict your test_df values using XGBoost.
```

```python
from xgboost import XGBClassifier
```

```python
xg=XGBClassifier()
```

```python
model=xg.fit(lda_train,target)
```

```python
model.score(lda_train,target)
```

```python
y_pred=model.predict(lda_test)
```
