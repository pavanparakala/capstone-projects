```python
x=data2.drop(["click_rate"],axis=1)
y=data2.click_rate
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
model_lin=lin.fit(X_train,y_train)
```

```python
model_lin.score(X_train,y_train)
```

```python
ypred_lin=model_lin.predict(X_test)
```

```python
model_lin.score(X_test,y_test)
```

```python
print("MAE: ",mean_absolute_error(y_test,ypred_lin))
print("rmse: ",np.sqrt(mean_absolute_error(y_test,ypred_lin)))
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)


redge = RidgeCV(alphas=(0, 1, 0.01), cv=cv, scoring='neg_mean_absolute_error')

#fit model
model_redge=redge.fit(X_train, y_train)
```

```python
model_redge.score(X_train, y_train)
```

```python
ypred_redge=model_redge.predict(X_test)
```

```python
model_redge.score(X_test,y_test)
```

```python
ypred_redge=model_redge.predict(X_test)
```

```python
model_redge.score(X_test,y_test)
```

```python
print("MAE: ",mean_absolute_error(y_test,ypred_redge))
print("rmse: ",np.sqrt(mean_absolute_error(y_test,ypred_redge)))
```
