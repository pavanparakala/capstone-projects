```python
df.Glucose.mean()
```

```python
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
```

```python
plt.boxplot(df["Glucose"])
plt.title("Glucose")
plt.show()
```

```python
plt.boxplot(df["BloodPressure"])
plt.title("BloodPressure")
plt.show()
```

```python
df.BloodPressure.describe()
```

```python
q1=df.BloodPressure.quantile(0.25)
q3=df.BloodPressure.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
lower,upper
```

```python
df=df[(df.BloodPressure>lower)&(df.BloodPressure<upper)]
```

```python
df.shape
```

```python
plt.boxplot(df["BloodPressure"])
plt.title("BloodPressure")
plt.show()
```

```python
plt.boxplot(df["SkinThickness"])
plt.title("SkinThickness")
plt.show()
```

```python
df.SkinThickness.mean()
```

```python
df.SkinThickness.value_counts()
```

```python
df["SkinThickness"]=df["SkinThickness"].replace(0,df.SkinThickness.mean())
```

```python
plt.boxplot(df["SkinThickness"])
plt.title("SkinThickness")
plt.show()
```

```python
plt.boxplot(df["BMI"])
plt.title("Bmi")
plt.show()
```

```python
q1=df.BMI.quantile(0.25)
q3=df.BMI.quantile(0.75)
iqr=q3-q1
lower=q1-1.5*iqr
upper=q3+1.5*iqr
lower,upper
```

```python
df=df[(df.BMI>lower)&(df.BMI<upper)]
```

8. Apply an appropriate classification algorithm to build a model. Compare various models with the results from KNN.

```python
X=ls.drop([8],axis="columns")
y=df.Outcome
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
k=range(1,60)
parameters={"n_neighbors": k}
```

```python
grid=GridSearchCV(knn,parameters,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
error=[]
for i in range(1,60):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(12,8))
plt.plot(range(1,60),error,color="black",linestyle="dashed")
```

```python
knn=KNeighborsClassifier(n_neighbors=19)
```

```python
model=knn.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
ypred=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
predict_proba1=model.predict_proba(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,ypred))
```
